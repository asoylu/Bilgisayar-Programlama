```python
import matplotlib.pyplot as plt
import numpy as np

x=[1,2,3,4,5]
y=[1,4,9,16,25]

plt.plot(x,y,"-b")
plt.axis([-1,8,0,20])
plt.title("y=x kare grafiği")
plt.xlabel("x ekseni")
plt.ylabel("y ekseni")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=[1,2,3,4,5]
y=[1,4,9,16,25]

plt.plot(x,y,color="red",linewidth="0.5")
plt.axis([0,6,0,20])
plt.title("y=x kare grafiği")
plt.xlabel("x ekseni")
plt.ylabel("y ekseni")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,2,100)

plt.plot(x,x,label="çizgisel")
plt.plot(x,x**2,label="kare")
plt.plot(x,x**3,label="küp")
plt.plot(x,x**4,label="karekare")



plt.xlabel("x ekseni")
plt.ylabel("y ekseni")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,2,100)

fig,axs=plt.subplots(3)
axs[0].plot(x,x,color="red")
axs[0].set_title("linear")

axs[1].plot(x,x**2,color="green")
axs[1].set_title("quadratic")

axs[2].plot(x,x**3,color="blue")
axs[2].set_title("kup")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,2,100)

fig,axs=plt.subplots(2,2)
fig.suptitle("Grafikler")

axs[0,0].plot(x,x,color="red")
axs[0,0].set_title("linear")

axs[0,1].plot(x,x**2,color="red")
axs[0,1].set_title("kare")

axs[1,0].plot(x,x**3,color="red")
axs[1,0].set_title("kup")

axs[1,1].plot(x,x**4,color="red")
axs[1,1].set_title("karekare")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(-10,9,20)
y=x**3
z=x**2
figure=plt.figure()
axes_cube=figure.add_axes([0.2,0.2,0.8,0.8])
axes_cube.plot(x,y,"b",label="cubic")
axes_cube.set_xlabel("x")
axes_cube.set_ylabel("y")
axes_cube.set_title("cube")
axes_cube.legend(loc=4)



axes_square=figure.add_axes([0.25,0.70,0.25,0.25])
axes_square.plot(x,y,"r",label="square")
axes_square.set_xlabel("x")
axes_square.set_ylabel("y")
axes_square.set_title("cube")
axes_square.legend(loc=4)

plt.savefig("Sekil-1.pdf")

plt.show()
```

```python
import matplotlib.pyplot as plt
yil=[2018,2019,2020,2021,2022]
oyuncu1=[8,9,10,7,12]
oyuncu2=[5,10,8,21,9]
oyuncu3=[8,10,14,20,5]

plt.plot([],[],color="y",label="oyuncu1")
plt.plot([],[],color="r",label="oyuncu2")
plt.plot([],[],color="b",label="oyuncu3")

plt.stackplot(yil,oyuncu1,oyuncu2,oyuncu3,colors=["y","r","b"])
plt.show()
```
